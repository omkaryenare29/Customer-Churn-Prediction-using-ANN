```python
# RowNumber is of no use
# CustomerId is of no use
# Surname is of no use

# So removing first three columns which are unnecessary
df = df.iloc[:,3:]
```

```python
# One Hot Encoding the 2 features - Geography and Gender
# lets use get_dummyies() method

df = pd.get_dummies(df,columns = ['Geography','Gender'],drop_first = True)
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_train.shape
```

```python
# Scaling of data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
x_train_scaled
```

```python
# Building the model

model = tf.keras.models.Sequential()

# lets take 1 input layer, 1 hidden layer(3 neurons,sigmoid) and 1 output layer(1 neuron, sigmoid)

# 1st hidden layer
model.add(tf.keras.layers.Dense(
    units = 3,
    activation = 'sigmoid', # relu gives better results in hidden layer
    input_dim = 11
))
#output layer
model.add(tf.keras.layers.Dense(
    units = 1,
    activation = 'sigmoid'
))
```

```python
model.summary()
```

```python
 model.compile(
     optimizer = 'adam',
     loss = 'binary_crossentropy',
)
```

```python
#training the model
history = model.fit(x_train_scaled,y_train,epochs = 30)
```

```python
model.layers[1].get_weights()
```

```python
y_log = model.predict(x_test_scaled)
```

```python
y_log = np.where(y_log>0.5,1,0)
```

```python
y_log.shape
```

```python
model.evaluate(x_test_scaled,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_log)
```
